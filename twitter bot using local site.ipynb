````python
# python -m http.server
import http.server
import cv2
import numpy as np
import webbrowser
import time
import pyautogui
import pytesseract
from matplotlib import pyplot as plt

def navigate_twitter():
    '''Navigate to the user page using webbroser'''
#     ie. 
    webbrowser.open(site_location)
    #test for page loading.. 
    while True:
        # whats happening is usually the last to load.. 
        logo = pyautogui.locateOnScreen('local_at_signal.png')
        if logo is not None:
            break
navigate_twitter()
# time.sleep(2)
myScreenshot = pyautogui.screenshot()
myScreenshot.save('screen.png')

img = cv2.imread('screen.png')   # Read the image
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
# take image from browser window..
plt.imshow(img_rgb)
````

````python
at_img = cv2.imread('local_at_signal.png', cv2.IMREAD_UNCHANGED)

sample_img = cv2.imread('screen.png', cv2.IMREAD_UNCHANGED)

result = cv2.matchTemplate(sample_img, at_img, cv2.TM_CCOEFF_NORMED)
plt.imshow(result, cmap='gray' )
plt.imsave("result.png", result)
plt.show()
# plt.savefig('foo.png')
min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(result)
w = at_img.shape[1]
h = at_img.shape[0]
cv2.rectangle(sample_img, max_loc, (max_loc[0] + w, max_loc[1] + h), (0,255,255), 2)
threshold = 0.89
yloc, xloc = np.where(result >= threshold)

for (x, y) in zip(xloc, yloc):
    cv2.rectangle(sample_img, (x, y), (x + w, y + h), (0,255,255), 2)
    cv2.rectangle(sample_img, (x, y), (x + w+1000, y + h), (255,0,0), 2)
    print(x,y)

plt.imshow(sample_img)
plt.show()

img= cv2.imread('screen.png')   # Read the image
mask = np.zeros(img.shape,np.uint8)
user_max_width= 500
mask = mask-1
for (x, y) in zip(xloc, yloc):
    mask[y:y+h,x:x+w+user_max_width] = img[y:y+h,x:x+w+user_max_width]
    
#     cv2.rectangle(sample_img, (x, y), (x + w, y + h), (0,255,255), 2)
#     cv2.rectangle(sample_img, (x, y), (x + w+300, y + h), (255,0,0), 2)
# plt.imsave("result.png", result)
# mask[y:y+h,x:x+w] = img[y:y+h,x:x+w]
plt.imshow(mask, cmap='gray' )
plt.imsave("mask.png",mask)


# get the text from the image using pytesserect ...
data = pytesseract.image_to_string(mask, lang='eng',config='--psm 6')
print(data)
````
